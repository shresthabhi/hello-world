# CIR Model to simulate changes in interest Rates and Liability hedging

The Cox Ingersoll Ross model is used to model the dynamics of interest rates and is an extension of the popular Vasicek Model to prevent negative interest rate

\begin{equation*}
dr = a(b-r_t)dt + \sigma \sqrt{r_t} dW_t
\end{equation*}


## Short Rate vs Annualized returns

In general, if we compound every 1/N of a year (i.e N times a year) we get:

$$\left(1+\frac{r}{N}\right)^{N}$$

```python
def inst_to_ann(r):
    """
    Convert Short rate to annual rate
    """
    return np.expm1(r)

def ann_to_inst(r):
    """
    Convert annual rate to Short Rate
    """
    return np.log1p(r)
```

$$ dr = a(b-r_t)dt + \sigma \sqrt{r_t} dW_t $$

```python
def cir(n_years = 10, n_scnearios = 10, a = 0.05, b = 0.03, sigma = 0.05, periods_per_year = 12, r_0 = None):
    """
    Implements the CIR model for interest rates
    dr=a(b−rt)dt+ σ(√rt)dWt
    """
    if r_0 is None:r_0 = b
    r_0 = ann_to_inst(r_0)
    dt = 1/periods_per_year
    
    num_steps = int(n_years*periods_per_year) + 1
    shock = np.random.normal(0, scale = np.sqrt(dt), size = (num_steps, n_scnearios) ) 
    rates = np.empty_like(shock)
    rates[0] = r_0
    
    for step in range(1, num_steps):
        r_t = rates[step - 1]
        dr_t = a*(b - r_t)*dt + sigma*np.sqrt(r_t)*shock[step]
        rates[step] = abs(r_t + dr_t)
    
    return pd.DataFrame(data = inst_to_ann(rates), index = range(num_steps))
    
```

```python
cir(sigma = 0.01).plot(figsize = (12,6), legend = False)
```
